```python
df = train_data[['Survived','Sex']]
df.head(20)
```

```python
# to find percentage of male and female survivors
male_survivor = 0
female_survivor = 0

male = len(df.loc[df.Sex == 'male'])
female = len(df.loc[df.Sex == 'female'])

for i in range(len(df.index)):
    if df.loc[i]['Survived'] == 1:
        if df.loc[i]['Sex'] == 'male':
            male_survivor += 1
        if df.loc[i]['Sex'] == 'female':
            female_survivor += 1

male_survivor /= male
female_survivor /= male

print('male survivor percent = '+str(male_survivor))
print('female survivor percent ='+str(female_survivor))
```
